Step 1. Import the necessary libraries

Step 3. Assign it to a variable called users 

Step 4. Discover what is the mean age per occupation 

Step 5. Discover the Male ratio per occupation and sort it from the most to the least 

Step 6. For each occupation, calculate the minimum and maximum ages 

Step 7. For each combination of occupation and sex, calculate the mean age 

Step 8. For each occupation present the percentage of women and men

```python
import pandas as pd
```

```python
# Importing the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

# Assigning it to a variable called users
users = pd.read_csv(url, sep='|')
users
```

```python
# mean age per occupation
mean_age_per_ocupation = users.groupby('occupation').agg({'age': ['mean']})
mean_age_per_ocupation
```

```python
# Male ratio per occupation and sorting it from the most to the least.

users['male'] = users.gender.apply(lambda x: True if x=='M' else False)
(users.groupby('occupation').male.sum() / users.groupby('occupation').gender.count()).sort_values(ascending = False).round(2)
```

```python
# calculating the minimum and maximum ages for each occupation.

min_max_age_per_occupation = users.groupby('occupation').agg({'age': ['min','max']})
min_max_age_per_occupation
```

```python
# calculating the mean age for each combination of occupation and sex.

mean_age_for_combination = users.groupby(['occupation', 'gender']).agg({'age': ['mean']}).round(2)
mean_age_for_combination
```

```python
#For each occupation, presenting the percentage of women and men.

gender_percentage = (users.groupby('occupation')['gender'].value_counts(normalize=True)*100).reset_index(name='percentage').round(2)
gender_percentage
```

```python
# importing the dataset from wind.txt and assigned to variable called data

data = pd.read_csv("E:\\Big Data Analytics\\Sem 1\\BDAT 1004_Data Programming\\GitHub\\Ritin.Patel.BDAT1004PS3\\wind.txt", delim_whitespace=True)
data
```

```python
# replace the first 3 columns by a proper datetime index. 

data["DateTime"] = pd.to_datetime(
    data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1)
)
data = data.drop(columns=["Yr","Mo","Dy"])
data
```

```python
# Function to fix the year issue

def fix_year(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return pd.Timestamp(year, x.month, x.day)

# Apply the fix_year function to the index
data.index = data.index.map(fix_year)
```

```python
# Setting the right dates as the index with datatype "datetime64[ns]"

data.index.astype("datetime64[ns]")
data
```

```python
# Compute how many values are missing for each location over the entire record.

number_of_missing_values = data.isnull().sum()
number_of_missing_values
```

```python
# Compute how many non-missing values there are in total.

total_non_missing_values = data.count().sum()
total_non_missing_values
```

```python
# Calculate the mean windspeeds of the windspeeds over all the locations and all the times. 

overall_mean_windspeed = data.mean().mean().round(2)
overall_mean_windspeed
```

```python
# Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days.

loc_stats = pd.DataFrame({
            'min': data.min(),
            'max': data.max(),
            'mean': data.mean().round(2),
            'std': data.std().round(2)
            })
loc_stats
```

```python
# Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

day_stats = pd.DataFrame({
            'min': data.min(axis=1),
            'max': data.max(axis=1),
            'mean': data.mean(axis=1).round(2),
            'std': data.std(axis=1).round(2)
            })
day_stats
```

```python
# Find the average windspeed in January for each location.

mean_january = data[data.index.month == 1].mean().round(2)
mean_january
```

```python
# Downsample the record to a yearly frequency for each location.

yearly_data = data.resample('Y').mean().round(2)
yearly_data
```

```python
# Downsample the record to a monthly frequency for each location.

monthly_data = data.resample('M').mean().round(2)
monthly_data
```

```python
# Downsample the record to a weekly frequency for each location.

weekly_data = data.resample('W').mean().round(2)
weekly_data
```

```python
# Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks. 
weekly_stats = data.iloc[1:].resample('W-MON').agg(['min', 'max', 'mean', 'std'])
weekly_stats_52 = weekly_stats.iloc[:52]
weekly_stats_52
```

# Question 7

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000. 
Don't forget to label your axes! 

```python
df_selected = us_marriage_divorce[['Year','Marriages_per_1000','Divorces_per_1000']]
val = df_selected.loc[df_selected.Year.isin([1900,1950,2000])]
val
```

```python
Year = val['Year'].tolist()
Year_str = []
for x in Year:
    Year_str.append(str(x))

Marriages = val['Marriages_per_1000'].tolist()
Divorces = val['Divorces_per_1000'].tolist()

plt.bar(Year_str,Marriages,0.3,label="Marriages")
plt.bar(Year_str,Divorces,0.3,label="Divorces")
       
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.title('Comparison of marriages and divorces in U.S.')
plt.show()
```
